```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('mall.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
plt.scatter(data[['INCOME']],data[['SPEND']])
plt.xlabel('Income')
plt.ylabel('Spend')
plt.show()
```

```python
km=KMeans(n_clusters=3)
```

```python
y_predicted=km.fit_predict(data[['INCOME','SPEND']])
```

```python
data['cluster']=y_predicted
data.head(10)
```

```python
km.cluster_centers_
```

```python
df1=data[data.cluster==0]
df2=data[data.cluster==1]
df3=data[data.cluster==2]
df1
```

```python
df2
```

```python
df3
```

```python
plt.scatter(df1.INCOME,df1['SPEND'],color='green')
plt.scatter(df2.INCOME,df2['SPEND'],color='red')
plt.scatter(df3.INCOME,df3['SPEND'],color='yellow')
plt.xlabel('INCOME')
plt.ylabel('SPEND')
plt.show()
```

```python
scaler=MinMaxScaler()
scaler.fit(data[['INCOME']])
data['INCOME']=scaler.transform(data[['INCOME']])
```

```python
scaler.fit(data[['SPEND']])
data['SPEND']=scaler.transform(data[['SPEND']])
data.head()
```

```python
km=KMeans(n_clusters=3)
```

```python
y_predicted=km.fit_predict(data[['INCOME','SPEND']])
```

```python
data['cluster']=y_predicted
data.head(10)
```

```python
df1=data[data.cluster==0]
df2=data[data.cluster==1]
df3=data[data.cluster==2]
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
plt.scatter(df1.INCOME,df1['SPEND'],color='green')
plt.scatter(df2.INCOME,df2['SPEND'],color='red')
plt.scatter(df3.INCOME,df3['SPEND'],color='yellow')
plt.xlabel('INCOME')
plt.ylabel('SPEND')
plt.show()
```

```python
sse=[]
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(data[['INCOME','SPEND']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('value of k')
plt.ylabel('value of SSE')
plt.plot(k_rng,sse)
plt.show()
```
